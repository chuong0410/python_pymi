# Bài 11_2

```python
from math import pi
```

```python
plt.style.use('seaborn')
plt.title('Bai 11_2')
x = np.linspace(-2 * pi, 2 * pi, 50)
y = np.sin(x)
z = np.cos(x)
t = np.tan(x)
plt.plot(x, y)
plt.plot(x, z)
plt.plot(x, t)
plt.xlabel("x")
plt.legend(['sin(x)', 'cos(x)', 'tan(x)'])
```
